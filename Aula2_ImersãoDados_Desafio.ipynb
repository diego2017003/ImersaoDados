```python
#importando os dados
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

dados = pd.read_csv(url_dados, compression = 'zip')
dados = dados.rename(columns=lambda x: "".join(x.split("-")))
dados
```

```python
dados.rename(columns={'droga': 'composto'}, inplace=True)
dados['composto'].value_counts().head()
```

```python
cod_compostos = dados['composto'].value_counts().index[0:5]
table_Composto = dados.query('composto in @cod_compostos')
table_Composto['composto'].unique()
```

#Desafio 1
---


```python
sns.set()
plt.figure(figsize=(8, 6))
ax = sns.countplot(x = 'composto', data=table_Composto,order=table_Composto['composto'].value_counts().index)
ax.set_title('Top 5 compostos')
plt.show()
```

#Desafio 2
---

```python
sns.set()
plt.figure(figsize=(12, 6))
ax = sns.countplot(x = 'composto', data=table_Composto,order=table_Composto['composto'].value_counts().index)
ax.set_title('Top 5 compostos',fontdict={'fontsize':25})
ax.tick_params(labelsize=20)
ax.set_ylabel("Quantidade",fontdict={'fontsize':25})
ax.set_xlabel("Composto",fontdict={'fontsize':25})
sns.set_theme(font='Calibri')
plt.show()
```

```python
g_means = dados.loc[:,'g0':'g771'].describe().T['mean']
plt.figure(figsize=(12,6))
g_MHist = sns.histplot(g_means,bins=50,color='green')
g_MHist.set_title('Média dos Genes',fontdict={'fontsize':20})
g_MHist.tick_params(labelsize=15)
g_MHist.set_ylabel("grupo de Médias",fontdict={'fontsize':20})
g_MHist.set_xlabel("Médias dos genes",fontdict={'fontsize':20})
sns.set_theme(font='Calibri')
plt.show()
```

```python
g_min = dados.loc[:,'g0':'g771'].describe().T['min']
plt.figure(figsize=(12,6))
g_minHist = sns.histplot(g_min,bins=50,color='blue')
g_minHist.set_title('Minimos dos Genes',fontdict={'fontsize':20})
g_minHist.tick_params(labelsize=15)
g_minHist.set_ylabel("Quantidade por valor Minimo",fontdict={'fontsize':20})
g_minHist.set_xlabel("Minimos dos genes",fontdict={'fontsize':20})
sns.set_theme(font='Calibri')
plt.show()
```

```python
g_max = dados.loc[:,'g0':'g771'].describe().T['max']
plt.figure(figsize=(12,6))
g_maxHist = sns.histplot(g_max,bins=50,color='blue')
g_maxHist.set_title('Máximos dos Genes',fontdict={'fontsize':20})
g_maxHist.tick_params(labelsize=15)
g_maxHist.set_ylabel("Quantidade por valor Máximo",fontdict={'fontsize':20})
g_maxHist.set_xlabel("Máximos dos genes",fontdict={'fontsize':20})
sns.set_theme(font='Calibri')
plt.show()
```

```python
import numpy as np
#aplicação de logaritmo para visualizar os dados em ordem de db não surte grande relevância em normalização para o DataSet 
plt.figure(figsize=(12,12))
plt.subplot(2,1,1)
g_min = dados.loc[:,'g0':'g771'].describe().T['min']
g_minL = -np.log(np.abs(g_min.values))
g_minHist = sns.histplot(g_minL,bins=50,color='blue')
g_minHist.set_title('Minimos dos Genes',fontdict={'fontsize':15})
g_minHist.tick_params(labelsize=15)
g_minHist.set_ylabel("Quantidade por valor Minimo",fontdict={'fontsize':10})
g_minHist.set_xlabel("Minimos dos genes",fontdict={'fontsize':10})
sns.set_theme(font='Calibri')

plt.subplot(2,1,2)
g_max = dados.loc[:,'g0':'g771'].describe().T['max']
g_maxL = np.log(g_max)
g_maxHist = sns.histplot(g_maxL,bins=50,color='blue')
g_maxHist.set_title('Máximos dos Genes',fontdict={'fontsize':15})
g_maxHist.tick_params(labelsize=15)
g_maxHist.set_ylabel("Quantidade por valor Máximo",fontdict={'fontsize':10})
g_maxHist.set_xlabel("Máximos dos genes",fontdict={'fontsize':10})
sns.set_theme(font='Calibri')
plt.show()
```

```python
plt.figure(figsize=(12,6))
c_means = dados.loc[:,'c0':'c99'].describe().T['mean']
c_meansHist = sns.histplot(c_means,bins=50,color='black')
c_meansHist.set_title('Média dos Cs',fontdict={'fontsize':15})
c_meansHist.tick_params(labelsize=15)
c_meansHist.set_ylabel("Quantidade por valor das Médias",fontdict={'fontsize':10})
c_meansHist.set_xlabel("Médias dos Cs",fontdict={'fontsize':10})
sns.set_theme(font='Calibri')
```

```python
#aplicação de log para alterar a distribuição
plt.figure(figsize=(12,6))
c_means = dados.loc[:,'c0':'c99'].describe().T['mean']
c_meansHist = sns.histplot((-np.log(abs(c_means))),bins=50,color='black')
c_meansHist.set_title('Média dos Cs',fontdict={'fontsize':15})
c_meansHist.tick_params(labelsize=15)
c_meansHist.set_ylabel("Quantidade por valor das Médias",fontdict={'fontsize':10})
c_meansHist.set_xlabel("Médias dos Cs",fontdict={'fontsize':10})
sns.set_theme(font='Calibri')
```

```python
plt.figure(figsize=(20,8))
plt.subplot(1,2,1)
sns.boxplot(y='g0',x='dose' , data=dados)
plt.subplot(1,2,2)
sns.boxplot(y='g0',x='tempo' , data=dados)
```

```python
plt.figure(figsize=(20,8))
plt.subplot(1,3,1)
sns.boxplot(y='g0',x='tratamento' , data=dados)
plt.subplot(1,3,2)
sns.boxplot(y='g1',x='tratamento' , data=dados)
plt.subplot(1,3,3)
sns.boxplot(y='g2',x='tratamento' , data=dados)
```

```python
plt.figure(figsize=(18,8))
teste = dados.loc[dados['tratamento']=='com_droga']
controle = dados.loc[dados['tratamento']=='com_controle']
plt.subplot(1,2,1)
sns.histplot(data=teste['g0'],bins=50).set_title("teste")
plt.subplot(1,2,2)
sns.histplot(data=controle['g0'],bins=50).set_title("controle")
```

```python
plt.figure(figsize=(24,8))
h24 = dados.loc[dados['tempo']==24]
h48 = dados.loc[dados['tempo']==48]
h72 = dados.loc[dados['tempo']==72]
plt.subplot(1,3,1)
sns.histplot(data=h24['g0'],bins=50).set_title("24 horas")
plt.subplot(1,3,2)
sns.histplot(data=h48['g0'],bins=50).set_title("48 horas")
plt.subplot(1,3,3)
sns.histplot(data=h72['g0'],bins=50).set_title("72 horas")
```

```python
plt.figure(figsize=(20,8))
D1 = dados.loc[dados['dose']=='D1']
D2 = dados.loc[dados['dose']=='D2']
plt.subplot(1,2,1)
sns.histplot(data=D1['g0'],bins=50,color='green').set_title("Dose D1")
plt.subplot(1,2,2)
sns.histplot(data=D2['g0'],bins=50,color='red').set_title("Dose D2")
```
